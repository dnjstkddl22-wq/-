## 파일_객체 = open(파일_이름, 파일_열기_모드)

```python
f = open("새파일.txt", 'w')
f.close()         
```

경로를 적을 떄는 / 또는 \ 또는 r""을 사용해야 한다. 

```python
f = open("새폴더//새파일.txt", 'w')
f.close()
```

## -.파일을 쓰기 모드로 열어 내용 쓰기

```python
f = open("새파일.txt","w", encoding="utf-8")


#반목문
for i in range(1,11):
    data = str(i) + "번째줄입니다.\n"

    print(data)
    f.write(data)


print("문자열 파일에 쓰기")
f.write("문자열 파일에 쓰기")
f.close()
```

## -.readline 함수 이용하기

```python
f = open("새파일.txt","r")
line = f.readline()
line2 = f.readline()
print(line)
f.close
```

## -.readlines 함수 사용하기

```python
f = open("새파일.txt",'r')
lines = f.readlines() #readlines() 파일 모든줄을 리스트로 반환한다.
print(lines)
f.close()
```

## -.read 함수 사용하기

```python
f = open("새파일.txt",'r')
lines = f.read() 
print(lines)
f.close()
```

## -.파일에 새로운 내용 추가하기

```python
f = open("my_log.txt",'a')

for i in range(11,20):
    data = f"{i}번째 줄입니다.\n"
    f.write(data)
    
f.close()
```

## -.with 문과 함께 사용하기

```python
with open ("foo.txt","w") as f:
    f.write("life is too short, you need python")
```
