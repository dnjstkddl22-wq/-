```python
import requests
import json
import random

apikey = '472C2678859F235C5AC0B162E969FEF9'
print("=== 끝말잇기 게임을 시작합니다! (끝내려면 '항복' 입력) ===")

# 컴퓨터가 방금 낸 단어를 기억할 빈 상자를 하나 만듭니다.
computer_last_word = "" 

# while True: 누군가 질 때까지 이 아래의 코드를 무한히 반복합니다!
while True:
    player = input('사용자: ')
    
    # 1. 사용자가 기권했는지 확인
    if player == '항복':
        print("컴퓨터: 저의 승리군요! 게임을 종료합니다. ")
        break # 무한 반복문을 깨고 게임을 끝냅니다.
        
    # 2. 끝말잇기 규칙 검사 (컴퓨터가 낸 단어가 있었을 경우에만)
    if computer_last_word != "":
        # 사용자의 첫 글자(player[0])가 컴퓨터 마지막 글자(-1)와 다르면?
        if player[0] != computer_last_word[-1]:
            print(f"컴퓨터: 땡! '{computer_last_word[-1]}' (으)로 시작해야 합니다. 다시 입력하세요!")
            continue # 아래 코드를 실행하지 않고, 다시 input()으로 올라갑니다.

    # 3. 컴퓨터의 턴 (API에서 단어 찾아오기)
    query = player[-1]
    url = f'https://opendict.korean.go.kr/api/search?certkey_no=8916&key={apikey}&target_type=search&req_type=json&part=word&sort=popular&start=1&num=100&pos=1&q={query}'
    
    response = requests.get(url)
    words = json.loads(response.text)
    
    list_a = []
    
    # 선생님의 코드 방식 그대로 적용 (대신 에러가 나지 않게 try-except를 씌웠습니다)
    for i in range(10):
        try:
            if words['channel']['item'][i]['sense']['word'][0] == query:
                # 기호(-, ^)를 깔끔하게 지워주는 선생님의 훌륭한 코드!
                clean_word = words['channel']['item'][i]['sense']['word'].replace('-', "").replace("^", "")
                list_a.append(clean_word)
        except:
            break # 검색 결과가 10개보다 적어서 에러가 날 것 같으면 안전하게 멈춥니다.
            
    # 4. 컴퓨터가 단어를 외치고, 그 단어를 기억하기
    if len(list_a) > 0:
        computer_word = random.choice(list_a)
        print('컴퓨터: ' + computer_word)
        
        # 다음 턴에 사용자가 제대로 잇는지 검사하기 위해 컴퓨터의 단어를 저장합니다.
        computer_last_word = computer_word 
    else:
        print('컴퓨터: 제가 졌습니다! 이을 단어가 없네요 ')
        break # 컴퓨터가 졌으니 무한 반복문을 깨고 게임을 끝냅니다.
```
